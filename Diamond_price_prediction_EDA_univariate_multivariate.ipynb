```python
plt.figure(figsize = (10,8))
sns.heatmap(data = df.corr(), cmap = 'RdBu',annot=True)
plt.title('Heatmap')
plt.show()
```

```python
## Univariate analysis for cut , color, clarity

plt.rcParams['figure.figsize']=23,5
plt.subplot(131)
plt.pie(df['cut'].value_counts().values,
        labels=df['cut'].value_counts().index,
        startangle=90,
        colors=['green', 'blue', 'red', 'cyan', 'pink'],
        explode=[0.05,0.05,0.05,0.05,0.2],
        shadow=True, autopct='%1.2f%%')
plt.subplot(132)

sns.countplot(df['color'], palette='winter', order=df['color'].value_counts().index)

plt.subplot(133)
df["clarity"].value_counts().plot.bar()
plt.xlabel('clarity types')
plt.ylabel('count')
plt.suptitle('Distribution of Cut, Color and Clarity of Gems')
plt.show()
```

```python
## Univariate analysis for depth, carat, price

plt.rcParams['figure.figsize']=20,5
plt.subplot(131)
sns.boxplot(df['depth'], color='royalblue')

plt.subplot(132)
sns.boxplot(df['carat'], color='red')

plt.subplot(133)
sns.boxplot(df['price'], color='green')

plt.suptitle('Distribution of depth carat and price')
plt.show()
```

```python
plt.rcParams['figure.figsize']=20,5
plt.subplot(131)
sns.boxplot(df['x'], color='royalblue')

plt.subplot(132)
sns.boxplot(df['y'], color='red')

plt.subplot(133)
sns.boxplot(df['z'], color='green')

plt.suptitle('Distribution of Length of the cubic zirconia')
plt.show()
```

```python
## Multivariate analysis

sns.pairplot(df, hue = 'cut')
plt.show()
```
